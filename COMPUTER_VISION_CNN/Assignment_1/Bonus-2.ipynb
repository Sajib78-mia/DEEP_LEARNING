```python
import matplotlib.pyplot as plt
import cv2

# Load the images using OpenCV (or use any image loading method)
image1 = cv2.imread('/content/drive/MyDrive/samples/cat.3570.jpg')
image2 = cv2.imread('/content/drive/MyDrive/samples/cat.394.jpg')
image3 = cv2.imread('/content/drive/MyDrive/samples/dog.6768.jpg')

# Convert the images from BGR to RGB
image1_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2_rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
image3_rgb = cv2.cvtColor(image3, cv2.COLOR_BGR2RGB)

# Create a figure with 3 subplots (3 rows, 1 column)
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(5, 15))

# Plot the first image in the first subplot
axes[0].imshow(image1_rgb)
axes[0].axis('off')  # Hide the axis for clean display
axes[0].set_title('Image 1')

# Plot the second image in the second subplot
axes[1].imshow(image2_rgb)
axes[1].axis('off')
axes[1].set_title('Image 2')

# Plot the third image in the third subplot
axes[2].imshow(image3_rgb)
axes[2].axis('off')
axes[2].set_title('Image 3')

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
