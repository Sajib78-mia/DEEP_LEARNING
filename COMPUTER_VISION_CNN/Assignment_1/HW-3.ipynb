```python
import cv2
import matplotlib.pyplot as plt

# Load the image using OpenCV
image_path = r'/content/drive/MyDrive/samples/dog.8643.jpg'
img_bgr = cv2.imread(image_path)

# Plot the image directly (without conversion) - colors will be wrong
plt.imshow(img_bgr)
plt.title('Image in BGR (Loaded with OpenCV)')
plt.axis('off')
plt.show()

# Convert the image from BGR to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Plot the image after conversion - colors will be correct
plt.imshow(img_rgb)
plt.title('Image in RGB (After Conversion)')
plt.axis('off')
plt.show()
```
