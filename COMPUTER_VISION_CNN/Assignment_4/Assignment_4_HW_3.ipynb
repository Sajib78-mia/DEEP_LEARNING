##  build mask(circle) and apply on the given image so that only the orange will be shown

```python
from PIL import Image, ImageDraw

# Load the new image
image_path_2 = "/content/thkoczih.png"
image_2 = Image.open(image_path_2)

# Create a mask for the circular area
mask_2 = Image.new("L", image_2.size, 0)
draw_2 = ImageDraw.Draw(mask_2)

# Draw a circle in the center of the image
center_2 = (image_2.size[0] // 2, image_2.size[1] // 2)
radius_2 = min(image_2.size) // 2
draw_2.ellipse((center_2[0] - radius_2, center_2[1] - radius_2, center_2[0] + radius_2, center_2[1] + radius_2), fill=255)

# Apply the mask to the image
result_2 = Image.new("RGBA", image_2.size)
result_2.paste(image_2, (0, 0), mask_2)
# Save the result
output_path_2 = "D:\Image_processing_CNN\Assignment_4.png"
result_2.save(output_path_2)

output_path_2

```
