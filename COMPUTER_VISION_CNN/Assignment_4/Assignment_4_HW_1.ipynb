## separate the orange from given image (single color)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('/content/thkoczih.png')  # Update with your image path
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert image to HSV color space
hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

# Define range for the color orange in HSV
lower_orange = np.array([5, 100, 100], dtype=np.uint8)
upper_orange = np.array([25, 255, 255], dtype=np.uint8)

# Create a mask for orange color
mask = cv2.inRange(hsv, lower_orange, upper_orange)

# Bitwise-AND mask and original image to extract the orange part
res = cv2.bitwise_and(image, image, mask=mask)

# Show the result
plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title('Mask')

plt.subplot(1, 3, 3)
plt.imshow(res)
plt.title('Isolated Orange')

plt.show()
```
